```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as dt

url = 'https://books.toscrape.com/' #'https://hungerstation.com/sa-ar'
url1 = url + 'catalogue/category/books/science_22/index.html'
response = requests.get(url1)
soup = BeautifulSoup(response.text, features="lxml")

# put the 'soup' object in a str method in order to parse its texts.
html_str = str(soup)

# '\w+.w+' finds all the one/more word characters after the symbol.
price = re.findall("£\w+.\w+", html_str)

# 'Title:True' because the anchor tag has title attribute, ant the title had texts that were too long to input here, hence,'True'.
title0 = soup.find_all("a", {"title": True})
print(price, title0)
#print(soup.find_all("h3"))
```

```python
# to slice the strings in book title in order to extract the main book title,find the start of the title string and index it '(.find)', then add the len of the start(e.g 7)
# to the .find position, and store the value.
titlestart = [str(t).find('title="') + len('title="') for t in title0]
titleend = [str(t).find('"', titlestart[i]) for i, t in enumerate(title0)]

# find the index position of the 'titlestart' and 'titleend' and give me the str of characters in between, starting from 'titlestart' but not including 'titleend'
title1 = [str(t)[titlestart[i]:titleend[i]] for i, t in enumerate(title0)]

print(title1)
```

```python
# using list comprehension to iterate through the prices and title while string them in a dictionary, to be later used in creating a dataframe.
frame = {"BookTitle":[t for t in title1],"BookPrice":[p for p in price]}
df = dt.DataFrame(frame)
df
```
